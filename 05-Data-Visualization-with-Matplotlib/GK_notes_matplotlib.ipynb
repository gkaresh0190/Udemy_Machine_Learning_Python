```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r-')
#plt.show() - similar to prinitng
#not in jupyter use plt.show()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) #(#rows, #cols, #plot number referring too)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,.8])
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig
```

```python
fig = plt.figure()

axes1 = fig.add_axes([.1,0.1,0.8,0.8])
axes2 = fig.add_axes([.2,.5,.4,.3])
#(from left, from top, width, heighth)

axes1.plot(x,y)
axes1.set_title('larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
#FIGURE SIZE and DPI
```

```python
fig = plt.figure(figsize=(3,2),dpi=100) #dpi = dots per inch or pixels per inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2)) #dpi = dots per inch or pixels per inch

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='x cubed')

ax.legend(loc=(0.1,0.1))
#ax.legend(loc=0) # if you add a legend then need to add another argument in ax.plot called 'label'
#loc=0 is top left, loc=1 is top right
```

```python
#Plot Appearance
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'orange', linewidth = 20, alpha=0.5, linestyle='-.',marker='o',markersize=20,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') 
#put in any color name or RGB hexcodes
# linewidth default =1, 20 is 20 times teh default or lw
# alpha is transparancey
#linestyle style of the line like dashes or dots, 1 dash is solid line
#markers mark the point x vs y: 'o' = circles, '1' three prong,'+'=+s
### possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple', linewidth = 2, linestyle='--') 

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
